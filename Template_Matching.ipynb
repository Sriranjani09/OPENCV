```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define paths to the video and template
video_path = r"C:\Users\srira\Documents\sem 5\JD.mp4"
template_path = r"C:\Users\srira\Documents\sem 5\JD.PNG"

# Load the video and the template image in grayscale
video = cv2.VideoCapture(video_path)
template = cv2.imread(template_path, cv2.IMREAD_GRAYSCALE)

# Check if the video and template image are loaded successfully
if not video.isOpened():
    print("Error: Could not open video.")
    exit()

if template is None:
    print("Error: Could not load template image.")
    exit()

# Get the template dimensions
w, h = template.shape[::-1]

# Set the threshold for matching accuracy (lower it to improve match chances)
threshold = 0.6  # Try starting with 0.6, and adjust if necessary

# Process each frame in the video
while True:
    ret, frame = video.read()
    
    if not ret:
        break
    
    # Convert the current frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Check if the template is larger than the frame; resize if necessary
    if template.shape[0] > gray_frame.shape[0] or template.shape[1] > gray_frame.shape[1]:
        template = cv2.resize(template, (min(template.shape[1], gray_frame.shape[1]), min(template.shape[0], gray_frame.shape[0])))
        w, h = template.shape[::-1]  # Update dimensions after resizing
    
    # Apply template matching
    result = cv2.matchTemplate(gray_frame, template, cv2.TM_CCOEFF_NORMED)
    
    # Get locations where the matching exceeds the threshold
    locations = np.where(result >= threshold)
    
    # Check if any matches were found
    if len(locations[0]) == 0:
        print("No matches found in this frame.")
    else:
        print("Match found!")
    
    # Draw rectangles on matched regions
    for pt in zip(*locations[::-1]):
        cv2.rectangle(frame, pt, (pt[0] + w, pt[1] + h), (0, 255, 255), 3)
    
    # Display the frame with the matched regions highlighted
    cv2.imshow("Template Matching", frame)
    
    # Press 'q' to exit the video display
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# Release video object and close all display windows
video.release()
cv2.destroyAllWindows()
```
